```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def collatz(n):
    cnt=0
    while n!=1:
        if n%2==0:
            n=n/2
        else:
            n=(3*n)+1
        cnt+=1
    return cnt
```

```python
collatz(7)
```

```python
num = list(range(1,21))
col_list = list(map(collatz,num))
xi = list(range(1,1001))
#num
#col_list
```

```python
plt.plot(num,col_list)
plt.show()
```
