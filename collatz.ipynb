```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def collatz(n):
    cnt=0
    while n!=1:
        if n%2==0:
            n=n/2
        else:
            n=(3*n)+1
        cnt+=1
    return cnt
```

```python
collatz(7)
```

```python
num = list(range(1,20001))
col_list = list(map(collatz,num))
#xi = list(range(1,10001))
#num
#col_list
df = pd.DataFrame(col_list)
num = pd.DataFrame(num)
#df.max()
df.idxmax()
```
